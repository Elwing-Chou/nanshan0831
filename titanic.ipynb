```python
# urllib->request.py 功能: urlretrieve(下載網址, 存放路徑)
from urllib.request import urlretrieve
url = "https://github.com/Elwing-Chou/nanshan0831/raw/main/data/train.csv"
urlretrieve(url, "train.csv")
url = "https://github.com/Elwing-Chou/nanshan0831/raw/main/data/test.csv"
urlretrieve(url, "test.csv")
```

```python
import pandas as pd
# pandas自訂兩種型態: DataFrame(有行有烈) Series(一行/一列)
train = pd.read_csv("train.csv", encoding="utf-8")
test = pd.read_csv("test.csv", encoding="utf-8")
```

```python
# train: 891 test: 418
# 資料多寡: 機器學習(幾千～幾萬) 深度學習(幾萬～幾十萬)
train
```

```python
combine = pd.concat([train, test], axis=0)
combine = combine.drop(["PassengerId", "Survived"], axis=1)
combine
```

```python
s = combine.isna().sum()
fil = s > 0
s[fil].sort_values(ascending=False)
```

輸入欄位要分兩種
1. 類別(固定數種): Pclass, Name(稱謂), Sex, Cabin(英文字母), Embarked
2. 數值: Age, SibSp, Parch, Ticket(多人跟我同號碼), Fare

1. 類別: 最常出現的
2. 數值: 中位數

```python
med = combine.median(numeric_only=True).drop("Pclass")
combine = combine.fillna(med)
```

```python
# value_counts()
counts = combine["Embarked"].value_counts()
combine["Embarked"] = combine["Embarked"].fillna(counts.idxmax())
```
